```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

```python
data = pd.read_csv("dataset/dataset_tcc_final.csv")
data["dataNotif"]=pd.to_datetime(data["dataNotif"], format='%d/%m/%Y')
data["dataNotif"]=data["dataNotif"].dt.strftime('%Y-%m')
data['totalObitos'] = data['saidaConfirmadaObitos'] + data['saidaSuspeitaObitos']
data
```

```python
df_resumo1 = data.groupby(['dataNotif']).sum().reset_index()
df_resumo1.head()
```

```python
df1 = df_resumo1[['dataNotif','totalObitos']]
df1.shape
```

```python
data = df1[:]["dataNotif"]
nObitos = df1[:]["totalObitos"]
```

```python
fig = plt.figure(figsize =(18, 5))
plt.bar(data, nObitos, color = 'blue')
plt.ylabel("No. Obitos")
plt.xlabel("Meses")
plt.title("Obitos por MÃªs")
plt.show()
```

```python
data4 = pd.read_csv("dataset/dataset_tcc_final.csv")
data4["dataNotif"]=pd.to_datetime(data4["dataNotif"], format='%d/%m/%Y')
data4["dataNotif"]=data4["dataNotif"].dt.strftime('%Y-%m')
data4['totalObitos'] = data4['saidaConfirmadaObitos'] + data4['saidaSuspeitaObitos']

df_resumo5 = data4.groupby(['no_bairro']).sum().reset_index()
df_resumo5.head()
```

```python
df_pre = df_resumo5[['no_bairro','totalObitos']]
df_pre.shape
```

```python
df_pre.sort_values(by=["totalObitos"], inplace=True, ascending=False)
df_ordered = df_pre.head(10)
df_ordered
```

```python
bairros = df_ordered[::-1]["no_bairro"]
nObitos = df_ordered[::-1]["totalObitos"]
```

```python
plt.barh(bairros, nObitos, color = 'orange')
plt.ylabel("Bairros")
plt.xlabel("No. Obitos")
plt.title("10 Bairros com Maior No. de Ã“bitos em SÃ£o Paulo")
plt.show()
```
